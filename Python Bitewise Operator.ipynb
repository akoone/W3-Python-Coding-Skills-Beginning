```python
#Bitewise OR Operator
# # The bitewise OR (|) Operator takes two equivalent length bit designs as boundries: if the two bits in the looked-at positions are 0, the next bit is zero. Of not, it is one. 

a=10
b=4

print('a | b=', a|b)
```

```python
# Bitewise XOR Operator
# The Python Bitewise XOR(^) Operatoralso known as the exclusive OR operator, is used to perform the OR operator, is used to perform the XOR operation on the two operands. XOR stands for 'exclusive or', and it returns true if and only if exactly one of the operands is true.


a=10
b=4

#print bitewise XOR operation
print('a^b=', a^b)
```

```python
# Bitewise NOT operator.

a=10

print('~a=', ~a)
```
