```python
#Local Variables
# Local variables are created inside a function and exist only during its execution.
# Tey can not can be accessed outside from the function.
# 
def greet():
    msg='Hello from inside the function!' 
    print(msg)

greet()
greet()
```

```python
# Creating local variable and try to access to outside of the function

def greet():
    msg='Hello!'
    print('Inside function:', msg)

greet()
print('Outside function:', msg)
```

```python
# Global Variables
# Global variables are declared outside all functions and can be accessed anywhere in the program, including inside function.

msg='Python is awesome!'

def display():
    print('Inside function:', msg)

display()
print('Outside function:', msg)
```

```python
# Use of Local and Global Variables

def fun():
    s='Me too.'
    print(s)

s='I love GeeksforGeeks'
fun()
print(s)
```

```python
# Modifying Global Variables Inside a Function
# Without global causes error
def fun():
    s+='GFG' # Error: Python thinks s is a local
    print(s)
s='I love GeeksforGeeks'
fun()
```

```python
# With global (works correctly)

s='Python is a great!'

def fun():
    global s
    s += 'GFG' # Modify global value
    print(s)
    s='Look for GeeksforGeeks Python Section' # Reassign global
    print(s)

fun()
print(s)
```

```python
# Global vs Local with Same Name


a=1 # Global Variable

def f():
    print('f():',a) #Uses global a

def g():
    a=2 #Local shadows global
    print('g():',a)

def h():
   global a
   a=3 # Modifies global a
   print('h():', a)

print('global:', a)
f()
print('global:', a)
g()
print('global', a)
h()
print('global:', a)
```
