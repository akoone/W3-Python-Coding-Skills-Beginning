```python
# Case Changing of Python String Method




text='geeKs For geEkS'
print('\nConverted String:')
print(text.upper())

print('\nConverted String:')
print(text.lower())

print('\nConverted String:')
print(text.title())

print('\nConverted String:')
print(text.swapcase())

print('\nConverted Strings:')
print(text.capitalize())

print('\nOriginal String:')
print(text)
```
