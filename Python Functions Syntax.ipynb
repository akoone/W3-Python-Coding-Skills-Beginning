```python
# Keyword Argument

#Explicitly specifying the parameter names, so the order doesn't matter


def student(fname, lname):
    print(fname, lname)
student(lname='Practice', fname='Geeks')
student(fname='Geeks', lname='Practice')
```

```python
# Positional Arguments
# In positional arguments, values are assigned to parameters based on their order in function call.

def nameAge(name, age):
    print('Hi, I am', name)
    print('My age is', age)

print('Case-1:')
nameAge('Suraj', 27)

print('\nCase-2:')
nameAge(27, 'Suraj')
```

```python
# Arbitrary Arguments
# *arg-Non-Keyword Arguments
# ** kwargs-Keyword Arguments

def myFun(*args, ** kwargs):
    print('Non-Keyword Arguments (*args):')
    for arg in args:
        print(arg)

    print('\nKeyword Arguments (**kwargs):')
    for key, value in kwargs.items():
        print(f'{key}=={value}')

# Function call with both types of arguments
myFun('Hey', 'Welcome', first='Geeks', mid='for', last='Geeks')
```

```python
# Anonymous Functions
# function without a name, lambda keyword is used to create anonymous functions:

def cube(x): return x*x*x #withoutlambda
cube_l=lambda x : x*x*x # with lambda

print(cube(7))
print(cube_l(7))
```

```python
# return statement in function
# The return statement ends a function and sends a value back to the calller. 

def square_value(num):
    """This function returns the square value of the entered number"""
    return num**2

print(square_value(2))
print(square_value(-4))
```

```python
# Recursive Functions
# A recursive function is a function that calls itself to solve a problem. It is commonly used in mathematical and divide-and-conquer problems. Always include a base case to avoid infinite recursion.

def factorial(n):
    if n==0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(4))
```
