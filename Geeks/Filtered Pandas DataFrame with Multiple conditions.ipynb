```python
# import module
import pandas as pd

# assign data
dataFrame = pd.DataFrame({'Name': [' RACHEL  ', ' MONICA  ', ' PHOEBE  ',
                                   '  ROSS    ', 'CHANDLER', ' JOEY    '],
                          
                          'Age': [30, 35, 37, 33, 34, 30],
                          
                          'Salary': [100000, 93000, 88000, 120000, 94000, 95000],
                          
                          'JOB': ['DESIGNER', 'CHEF', 'MASUS', 'PALENTOLOGY',
                                  'IT', 'ARTIST']})

# display dataframe
display(dataFrame)
```

```python
# filter dataframe


display(dataFrame.loc[(dataFrame['Salary']>=100000)& (dataFrame['Age']<40)& (dataFrame['JOB'].str.startswith('D')),['Name','JOB']])
```

```python
# Filter Pandas Dataframe using NumPy


import pandas as pd
import numpy as np




# Filter dataframe


filtered_values=np.where((dataFrame['Salary']>=100000)&(dataFrame['Age']<40)& 
(dataFrame['JOB'].str.startswith('D')))
print(filtered_values)
display(dataFrame.loc[filtered_values])
```

```python
# Filter Pandas Dataframe Using Query (eval and query works only with columns)



display(dataFrame.query('Salary<=100000 & Age<40& JOB.str.startswith("C").values'))
```

```python
# Pandas Boolean indexing mu;tiple conditions standard way ('Boolean indexing' works with values in a column only)

display(dataFrame[(dataFrame['Salary']>=100000)&(dataFrame['Age']<40)&
dataFrame['JOB'].str.startswith('P')][['Name', 'Age','Salary']])
```

```python
# Eval multiple conditions ('eval' and 'query' works only with columns)

# filterdataframe


display(dataFrame[dataFrame.eval('Salary<=100000 & (Age<40)&JOB.str.startswith("A").values')])
```
