```python
# Indexing and Selecting Data with Pandas

# Indexing Data Using the [] Operator


import pandas as pd

# Recreate the DataFrame manually
data = {
    'Name': ['Avery Bradley', 'Jae Crowder', 'John Holland', 'R.J. Hunter', 'Jonas Jerebko'],
    'Team': ['Boston Celtics'] * 5,
    'Number': [0.0, 99.0, 30.0, 28.0, 8.0],
    'Position': ['PG', 'SF', 'SG', 'SG', 'PF'],
    'Age': [25.0, 25.0, 27.0, 22.0, 29.0],
    'Height': ['6-2', '6-6', '6-5', '6-5', '6-10'],
    'Weight': [180.0, 235.0, 205.0, 185.0, 231.0],
    'College': ['Texas', 'Marquette', 'Boston University', 'Georgia State', None],
    'Salary': [7730337.0, 6796117.0, None, 1148640.0, 5000000.0]
}

df = pd.DataFrame(data)
df.set_index('Name', inplace=True)

# Save to CSV
df.to_csv('boston_celtics_data.csv')

print("CSV file saved successfully.")
```

```python
# Selecting a Single Colummn


import pandas as pd


data=pd.read_csv('boston_celtics_data.csv', index_col='Name')
print('Dataset')
display(data.head(5))

first=data['Age']
print('\nSingle Column selected from Dataset')
display(first.head(5))
```

```python
# Selecting Multiple Columns 

first = data[['Position','Age','College','Salary']]
print('\nMultiple Columns selected from Dataset')
display(first.head(5))
```

```python
# Indexinf with .loc[]


#  .loc[] function is used for label-based indexing. It allows us to access rows and columns by their labels. Unlike the indexing operator, it can select subset of rows and columns simultaneously which offers flexibility in data retrival..

# Selecting a Single Row by Label


import pandas as pd
data=pd.read_csv('boston_celtics_data.csv', index_col='Name')

row=data.loc['Avery Bradley']
print(row)
```

```python
# Selecting Multiple rows by Label


rows=data.loc[['Avery Bradley','R.J. Hunter']]
print(rows)
```

```python
# Selecting Specifiv Rows and Columns


selection=data.loc[['Avery Bradley','R.J. Hunter'],['Team','Number','Position']]
print(selection)
```

```python
# Selecting all Rows and Specific Columns
# We can sellect all rows and colums by using [:] operation



all_rows_specific_columns=data.loc[:,['Team','Position','Salary']]
print(all_rows_specific_columns)
```

```python
# Indexing with .iloc[] is position based indexing, it allows us to access rows and columns by their intiger positions. 
# Selecting a Single Row by Position


import pandas as pd
data=pd.read_csv('boston_celtics_data.csv', index_col='Name')
row=data.iloc[3]
print(row)
```

```python
# Selecting Multiple Rows by Position

rows = data.iloc[[0,3,4]]
print(rows)
```

```python
# Selecting Specific Rows and Columns by Position

selection=data.iloc[[3,4],[1,2]]
print(selection)
```

```python
# Selecting all rows and Specific columns


selection=data.iloc[:,[1,2]]
print(selection)
```

```python
# Other Useful Indexing Methods


# .head() returns the first n row of a DataFrame

print(data.head(5))
```

```python
# .tail(): Returns the last n rows of a DataFrame


print(data.tail(5))
```

```python
# .at[]: Access a single value for a row/column laber pair


value=data.at['Avery Bradley', 'Age']
print(value)
```

```python
# .query(): Query the DataFrame using a boolean expression


result=data.query('Age>=25 and College=="Texas"')
print(result)
```
