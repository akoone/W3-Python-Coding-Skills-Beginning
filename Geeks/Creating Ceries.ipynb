```python
# Creating a Series from NumPy Array


import pandas as pd
import numpy as np


data=np.array(['g','e','e','k','s'])


ser=pd.Series(data)


print(ser)
```

```python
# Crwating a Series from a List

import pandas as pd

data_list=['g','e','e','k','s']

ser=pd.Series(data_list)


print(ser)
```

```python
# Creating a Series from Dictionary




import pandas as pd

data_dict={'Geeks':10,'for':20,'geeks':30}


ser=pd.Series(data_dict)


print(ser)
```

```python
# Creatong a Series Using NumPy functions

import numpy as np
import pandas as pd


ser=pd.Series(np.linspace(1,10,5))

print(ser)

```

```python
# Creating a Series using range()



import pandas as pd
ser=pd.Series(range(5,15))
print(ser)
```

```python
# Creating a Series Using List Comprehension

import pandas as pd
ser=pd.Series(range(1,20,3), index=[x for x in 'abcdefg'])
print(ser)
```
