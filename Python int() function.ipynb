```python
# Python int() Function for Custom Objects
class Number:
    value=7
    def __int__(self):
        return self.value
        #this is a magic method //dunder method//
        data=Number()          
        print('number=', int(data))
```

```python
class Number:
    value=7

    def __index__(self):
        return self.value

data=Number()
print('number=', int(data))
```
