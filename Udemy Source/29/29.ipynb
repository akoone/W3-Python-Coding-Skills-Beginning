```python

numbers = [1,35,12,24,31,51,70,100]
 
def count(numbers):
    numbers=sorted(numbers)
    tot = 0 # How many numbers are less than 20:
    
    
    while numbers[tot] < 20: # The loop continuous as long as the number at position tot is less that 20:
        tot += 1 # Used loops to count things
    return tot
 
count(nums)
```
