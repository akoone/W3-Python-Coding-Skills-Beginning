```python
# Functions can provide tuples as return values

def square_info(x):
    A=x**2
    P=4*x
    print("Area and Perimeter:")
    return A,P
square_info(3)
```

```python
def rectangle_info(x,y):
    Length=x*y
    Width=2*(y+x)
    print("Length and Width:")
    return Length, Width


rectangle_info(2,10)
```
