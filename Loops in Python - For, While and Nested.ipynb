```python
# While Loop
# a while loop is used to execute a block of statement repeatedly intill a gicen condition is satisfied.
# When the condition becomes false, the line immediately after the loop in the program is executed.
# Indent - to start a line of text further in from the margin than the other lines.
# Increment - increase or adding


cnt=0
while (cnt<3):
    cnt=(cnt+1)
    print('Hello Geeks')
```

```python
# Using else statement with while loop
# else clouse is only executed when our while condition becomes false..
# if we break out of the loop or if an exception is rised then it won't be executed
# 


cnt=0
while (cnt<3):
    cnt=(cnt+1)
    print('Hello Greek')
else:
    print('In else block')
    
```

```python
# Nested Loops 

from __future__ import print_function
for i in range(1,50):
    for j in range(i):
        print(i, end='')
    print()
```

```python
# Loop control statements
# Continue statement

for letter in 'geeksforgeeks':
    if letter =='e' or letter=='s':
        continue
    print('Current letter:', letter)
```

```python
# Break the statement

for letter in 'geeksforgeeks':
    if letter =='e' or letter=='s':
        break
print('Current Letter:', letter)
```

```python
#pass statement

for letter in 'geeksforgeeks':
    pass
print('Pass Letter:', letter)


```

```python
fruits = ['apple', 'orange', 'kiwi']
iter_obj=iter(fruits)
while True:
    try:
        fruit=next(iter_obj)
        print(fruit)
    except StopIteration:
        break
```
