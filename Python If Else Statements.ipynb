```python
# Logical Operators with if else

age=25
exp=10

if age >23 and exp>8:
    print('Eligible')
else:
    print('Not Elegible')
```

```python
# Nested if else statement
i=10
if i==10:

    if i<15:
        print('i is smaller than 15')

        if i<12:
            print('i is smaller than 12 too')
        else:
            print('i is greater than 15')

else:
    print('i is not equal to 10')
```

```python
# if elif else statement

i=25

if i==10:
    print("i is 10")

elif i==15:
    print('i is 15')

elif i==20:
    print('i is 20')
elif i==25:
    print('i is equal to 255')

else:
    print('i is not present')
```

```python
# if else Statement in Python
x=3
if x==4:
    print("Yes")
else:
    print('No')
```

```python
# Nested if else chain

letter='A'

if letter=='B':
    print('letter is B')
else:
    if letter=='C':
        print('letter is C')

    else:
        if letter=='A':
            print('letter is A')

        else:
            print('letter is not A,B,C')
```

```python
# if statement can also be checked inside other statement. This conditional statement is called nested if statement.
# This means that inner if condition will be checked only if outer if condition is true and by this, we can see multiple conditions to be satisfied.
#  
a=10

if a>5:
    print('Bigger than 5')
if a<=15:
    print('Between 5 and 15')
```

```python
# use elif within nested if statement

x=10
y=5

if x>5:
    if y>5:
        print('x is greater than 5')
    elif y==5:
        print('x is greater than 5 and y is 5')

    else: 
        print('x is greater than 5 and y is less than 5')
```
