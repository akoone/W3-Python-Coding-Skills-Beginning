```python
#Recursion is a programming technque where a function calls itself either directly or indirectly to solve a problem by breaking it into smaller, simpler subproblems.
# In python, recursion is especially useful for problems that can be divided into identical smaller tasks, such as mathematical calculations, tree traverse or divide-and-conquer algorithms.
# Recursive funvtion contains two key parts:
# Base Case: The stopping condirion that prevents infinite recursion.
# Recursive Case: The part of the functionwhere it calls itself with modified parameters.


# Factorial Calculation
def factorial(n):
    if n==0: # Base case
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(0))
```

```python
# Fibanacci Sequence
def fibonacci(n):
    if n==0:
        return(0)
    elif n==1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(10))
    
```

```python
# Types of Recursion in Python
# 
# Tail-Recursion
# Non-Tail Recursion
# 
# Tail-Recursion: The recursive call is the last thing the function does, so nothing happens after it returns. Soe lanuages can optimize this to work like a loop, saving memory.
#  Non-Tail Recursion: The function does more after the recursive call returns, so it can not be optimized into a loop.

def tail_fact(n, acc=1):
    #Base case
    if n==0:
        return acc
    # Tail recrusive call with an accumulator
    else:
        return tail_fact(n-1, acc * n)
    
def nontail_fact(n):
        # Base case
        if n==1:
            return 1
        # Non-tail recursive call because the multiplication happens after the call
        else:
            return n * nontail_fact(n-1)
        
        # Example Usage
print(tail_fact(5))
print(nontail_fact(5))
```
