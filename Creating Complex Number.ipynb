```python
# Creating Complex Numbers

c1=complex(3, 4)
c2=complex(5)
c3=complex()

print(c1)
print(c2)
print(c3)
print(c1.real)
print(c1.imag)
```

```python
c1 = complex(3, 4)
print(c1)

c2 = complex(2.5, 3.7)
print(c2)
```

```python
# Arthimetic of Complex Numbers

c1=complex(4, 3)
c2=complex(2, -5)


print(c1+c2)
print(c1-c2)
print(c1*c2)
print(c1/c2)

print(c1.real)
print(c2.imag)
```
